# Limits in Calculus

In calculus, **limits** are a fundamental concept that describe how a function behaves as its input approaches a particular value, or point. 
## What is a Limit?

A limit of a function $f(x)$ as $x$ approaches a value $a$ is the value that $f(x)$ gets closer and closer to as $x$ gets closer to $a$. This is mathematically expressed as:


$\lim_{x \to a} f(x) = L$

where $L$ is the limit, the value that $f(x)$ approaches as $x$ approaches $a$.





Consider the limit expression as \(x\) approaches 1:

$
\lim_{x \to 1} \frac{x^2 - 1}{x - 1}
$

We can simplify this expression by factoring the numerator:

$
\lim_{x \to 1} \frac{(x-1)(x+1)}{x - 1}
$

Canceling the $(x-1)$ terms valid because we are considering the limit where $x \neq 1$:

$
\lim_{x \to 1} (x+1) = 2
$




```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10,5, 50)
y = ((x**2) - 1) / (x - 1)
gap_min = 0.9
gap_max = 1.9
mask = (x > gap_min) & (x < gap_max)
x[mask] = np.nan
y[mask] = np.nan
plt.plot(x, y, 'b-')  
plt.title('Plot of $(x^2 - 1) / (x - 1)$ with a Gap at $x = 1$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

- The Limit of $f(x)$ as $x$ approaches $a$ is $L$
 $\lim_{x\to a}f(x)=L$
- If for every $\epsilon > 0 $ there is a $\delta > 0$ such thatn if $ 0<|x-a|<\delta$  then  $|f(x)-L|<\epsilon$

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(8, 6))
X = np.linspace(-10, 10, 100)
y = 2 * X + 3
plt.plot(X, y)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.axis([-1, 10, -1, 15])
plt.xticks(np.arange(-1, 11, 1))
plt.yticks(np.arange(-1, 15, 1))
X_coords, y_coords= zip((1,0), (1,5), (-1,5))
plt.plot(X_coords,y_coords)
X_coords, y_coords = zip((2,0), (2,7), (-1,7))
plt.plot(X_coords,y_coords)
X_coords, y_coords = zip((3,0), (3,9), (-1,9))
plt.plot(X_coords,y_coords)
plt.text(1.9,-.5,"$a$",fontsize=16,color='g')
plt.text(0.5,-.5,"$a-\\delta$",fontsize=14,color='orange')
plt.text(2.5,-.5,"$a+\\delta$",fontsize=14,color='red')
plt.text(1.,3,"$-\\delta-$",fontsize=14,color='b')
plt.text(2,3,"$-\\delta-$",fontsize=14,color='b')
plt.plot([0.2,0.2],[5.1,6.9], 'b-')
plt.plot([0.2,0.2],[7.1,8.9], 'b-')
plt.text(0.3,6,"$\\epsilon$",fontsize=14,color='b')
plt.text(0.3,8,"$\\epsilon$",fontsize=14,color='b')
plt.text(-.5,7.1,"$a$",fontsize=16,color='g')
plt.text(-.9,5.1,"$a-\\delta$",fontsize=14,color='orange')
plt.text(-.9,9.1,"$a+\\delta$",fontsize=14,color='red')
plt.show()

```

# Limit Laws

## Basic Limit Laws

Here are some of the fundamental limit laws used in calculus:

1. **Sum Law**:$\lim_{x \to a} [f(x) + g(x)] = \lim_{x \to a} f(x) + \lim_{x \to a} g(x)$

2. **Difference Law**: $\lim_{x \to a} [f(x) - g(x)] = \lim_{x \to a} f(x) - \lim_{x \to a} g(x)$

3. **Product Law**: $\lim_{x \to a} [f(x) \cdot g(x)] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x)$

4. **Quotient Law**:$\lim_{x \to a} \frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)}, \quad \text{if } \lim_{x \to a} g(x) \neq 0$

5. **Constant Multiple Law**: $\lim_{x \to a} [c \cdot f(x)] = c \cdot \lim_{x \to a} f(x)$

6. **Power Law**: $\lim_{x \to a} [f(x)^n] = (\lim_{x \to a} f(x))^n$

## Special Limits

- **Exponential Functions**:
  $\lim_{x \to 0} e^x = 1$

- **Natural Logarithm**:
  $\lim_{x \to 1} \ln(x) = 0$




```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 400)
f = np.sin(x)
g = np.cos(x)
h = f + g
plt.figure(figsize=(10, 6))
plt.plot(x, f, label='f(x) = sin(x)')
plt.plot(x, g, label='g(x) = cos(x)')
plt.plot(x, h, label='h(x) = sin(x) + cos(x)', linestyle='--')
plt.title('Basic Limit Laws: Sum Law')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.legend()
plt.show()
```


## Limits Involving Infinity

1. **Limit as x approaches positive infinity**:
   $\lim_{x \to \infty} f(x)$
   This limit describes what happens to $f(x)$ as $x$ becomes very large positively.

2. **Limit as x approaches negative infinity**:
   $\lim_{x \to -\infty} f(x)$
   This examines the behavior of $f(x)$ as $x$ becomes very large negatively.

### Examples

- **Exponential Decay**:
  $\lim_{x \to \infty} e^{-x} = 0$

- **Polynomial Growth**:
  $\lim_{x \to \infty} x^2 = \infty$

## Asymptotes

Asymptotes are lines that a graph of a function approaches but never touches. They can be horizontal, vertical, or oblique (slant).

1. **Horizontal Asymptotes**:
   If $\lim_{x \to \infty} f(x) = b$ or $\lim_{x \to -\infty} f(x) = b$, then $y = b$ is a horizontal asymptote.

2. **Vertical Asymptotes**:
   If $\lim_{x \to a} f(x) = \infty$ or $\lim_{x \to a} f(x) = -\infty$, then $x = a$ is a vertical asymptote.

3. **Oblique Asymptotes**:
   If the degree of the numerator is one more than the degree of the denominator in a rational function, the function may have an oblique asymptote, which can be found by polynomial long division.

### Example

- **Rational Function**:
  $\lim_{x \to 1} \frac{1}{x-1} = \infty$
  Indicates a vertical asymptote at $x = 1$.




```python
x = np.linspace(1, 100, 400)
y = x**2 / (x + 1)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = x^2 / (x + 1)')
plt.title('Limits Involving Infinity')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0, 110)
plt.grid(True)
plt.legend()
plt.show()
```

```python

x = np.linspace(-2, 2, 400)
x = x[x != 1] 
y = 1 / (x - 1)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = 1 / (x - 1)')
plt.title('Asymptotes: Vertical Asymptote at x = 1')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-50, 50) 
plt.grid(True)
plt.legend()
plt.show()
```

# Indeterminate Forms in Calculus



## Common Indeterminate Forms

Here are some of the most common indeterminate forms:

- **$0/0$**: Both the numerator and the denominator approach zero.
  Example: $\lim_{x \to 0} \frac{\sin(x)}{x}$

- **$/$**: Both the numerator and the denominator approach infinity.
  Example: $\lim_{x \to \infty} \frac{x^2}{x}$

- **$0  $**: One part approaches zero, and the other part approaches infinity.
  Example: $\lim_{x \to 0} x \cdot \ln(x)$

- **$ - $**: Differences where both terms approach infinity.
  Example: $\lim_{x \to \infty} (x - \sqrt{x^2 + 2x})$

- **$1^$**: A base approaching one raised to the power approaching infinity.
  Example: $\lim_{x \to 0} (1 + x)^{1/x}$

- **$0^0$**: Both the base and the exponent approach zero.
  Example: $\lim_{x \to 0} x^x$

- **$^0$**: The base approaches infinity while the exponent approaches zero.
  Example: $\lim_{x \to \infty} x^{1/x}$



```python
x = np.linspace(-10, 10, 400)
x = x[x != 0]  
y = np.sin(x) / x
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = sin(x) / x')
plt.title('Indeterminate Forms: sin(x) / x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
