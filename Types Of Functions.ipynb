## 1.Linear Functions

<p>Functions that produce only lines are called <b>linear functions</b>. A linear function can be represented by the equation of a straight line, typically in the form <em>y = mx + b</em>, where <em>m</em> is the slope of the line and <em>b</em> is the y-intercept. These functions are characterized by their constant rate of change and produce graphs that are straight lines.</p>

<p> let's look at a line with <em> slope </em> as 3, <em> y-intercept </em> as 2 </p> 

```python
import numpy as np 
import matplotlib.pyplot as plt
X = np.linspace(-8, 8, 400)
y = 3*X + 2
plt.axhline(y=0, color= "k")
plt.axvline(x=0, color= 'k')
plt.grid()
plt.plot(X,y,label='$3X+2$')
plt.axis([-6,6,-15,15])
plt.legend()
```

## 2. Polynomial Functions

 <p>Polynomial functions are mathematical expressions of the form <b><em>P(x) = a<sub>n</sub>x<sup>n</sup> + ... + a<sub>1</sub>x + a<sub>0</sub></em></b> , where <em>x</em> is the variable and <em>a<sub>n</sub></em> through <em>a<sub>0</sub></em> are constants. The highest power of <em>x</em>, <em>n</em>, determines the degree of the polynomial.</p>
    <p>These functions are continuous, have no breaks or sharp corners, and can intersect the horizontal axis at points called roots.</p>

<p><strong>P(x) = 3x<sup>3</sup> - 2x<sup>2</sup> + 5x - 8</strong></p>
    <h4>Explanation:</h4>
    <ul>
        <li><strong>Degree:</strong> The highest power of <em>x</em> in this polynomial is 3, making it a cubic polynomial.</li>
        <li><strong>Coefficients:</strong> The coefficients of the polynomial are 3 for <em>x<sup>3</sup></em>, -2 for <em>x<sup>2</sup></em>, 5 for <em>x</em>, and -8 as the constant term.</li>
        <li><strong>Variable:</strong> <em>x</em> is the variable in this polynomial.</li>
    </ul>

```python
X = np.linspace(-8, 8, 1000)
y = 3*(X**3) + 2*(X**2)+ 5*X - 8  
plt.axhline(y=0, color= "k")
plt.axvline(x=0, color= 'k')
plt.grid()
plt.plot(X, y, 'b-', label='$3X^3 + 2X^2 + 5X - 8$')
plt.axis([-8,8,-1500,1500])
plt.legend()
```

## 3.Root Functions

* A root function is generally expressed as: **$f(x)=\sqrt[n]{x}$**, where **$\sqrt[n]{x}$** denotes the nth root of **$x$**. The most frequently encountered root function is when **$n = 2$**, which yields the square root of **$x$**. For example, **$f(x)=\sqrt[]{x}$**.

### Example Calculation:

**Input:**  
$ x = 81 $

**Output:**  
$ f(81) = \sqrt[4]{81} = 3 $

In this example, taking the fourth root of 81 yields 3. This result reflects the property that  $3^4 = 81$.


* Lets look at example $3X^3 + \sqrt[4]{X} + 5X - 8$ to plot on a graph. 

```python
X = np.linspace(-8, 8, 100)
y = 3*(X**3) + X**(0.25)+ 5*X - 8 
plt.axhline(y=0, color= "k")
plt.axvline(x=0, color= 'k')
plt.grid()
plt.plot(X, y, 'g-', label='$3X^3 + X^{0.25} + 5X - 8$')
plt.legend()
plt.axis([-0.5,8,-100, 1600])
plt.show()
```

## 4. Exponential Functions

Exponential functions are mathematical functions of the form $ f(x) = a^x $, where:
- $ a $ is the base of the exponential function,
- $ a $ is a positive real number,
- $ a \neq 1 $,
- $ x $ is the exponent, which can be any real number.

### Characteristics of Exponential Functions

- The base, $ a $, is constant and positive.
- The function grows rapidly as $ x $ increases if $ a > 1 $.
- The function decays as $ x $ increases if $ 0 < a < 1 $.
- The graph of $ f(x) = a^x $ always passes through the point $ (0, 1) $ because any non-zero number raised to the power of 0 is 1.

### Common Bases

- **Base e**: When the base is Euler's number (approximately 2.71828), the function is called the natural exponential function, denoted as $ e^x $ or
  $ \exp(x) $.
- **Base 10**: This is expressed as $ 10^x $.


### Graph of Exponential Functions

The graph of an exponential function $ y = a^x $ depends on the value of $ a $:
- If $ a > 1 $, the graph increases (or grows) exponentially from left to right.
- If $ 0 < a < 1 $, the graph decreases (or decays) exponentially from left to right.




### $ a > 1 $, the graph increases
- Let's Look at this Graph  with example $ 1000. e^x $

```python
X = np.linspace(-8, 8, 100)
y =  1000*np.exp(0.48*X)
plt.axhline(y=0, color= "k")
plt.axvline(x=0, color= 'k')
plt.grid()
plt.plot(X, y, 'r-', label='$1000. e^{0.48X}$')
plt.legend()
plt.title("Increasing Graph")
plt.show()
```

### $ 0 < a < 1 $, the graph decreases
- Let's Look at this Graph  with example $ 1000.- e^x $

```python
X = np.linspace(-8, 8, 100)
y =  1000*-np.exp(0.48*X)
plt.axhline(y=0, color= "k")
plt.axvline(x=0, color= 'k')
plt.grid()
plt.plot(X, y, 'r-', label='$1000. -e^{0.48X}$')
plt.legend()
plt.title("Decreasing Graph")
plt.show()
```

## 5. Logarithmic Functions

Logarithmic functions are the inverse functions of exponential functions. The logarithm of a number is the exponent to which another fixed value, the base, must be raised to produce that number. For example, in the equation $ y = \log_b(x) $, $ y $ is the power to which $ b $ must be raised to yield $ x $.

### Definition

The logarithmic function can be defined as follows:

$ y = \log_b(x) \iff b^y = x $

where:
 * $ b $ is the base and $ b > 0 $,
 * $ b \neq 1 $,
 * $ x > 0 $.

### Common Bases

 * **Base 10**: Logarithms with base 10 are called common logarithms. Denoted as $ \log(x) $.
 * **Base e or $\ln(x)$**: Logarithms with base $ e $ are called natural logarithms. Denoted as $ \ln(x) $.

### Graphs

The graph of a logarithmic function $ y = \log_b(x) $ is a curve that passes through $ (1,0) $ on the x-y plane, since $ b^0 = 1 $, and it approaches the y-axis asymptotically but never touches or crosses it.




```python
X = np.linspace(0.1, 10, 400)
y = np.log10(X)
plt.plot(X, y, label=' $ \\log_{10}(x)$ ')
plt.axhline(0, color='k')  
plt.axvline(0, color='k')  
plt.title('Graph of the Logarithmic Function $ y = \\log_{10}(x) $')
plt.grid(True)
plt.legend()
plt.show()
```

## 6. Trigonometric Functions

Trigonometric functions are fundamental in mathematics, particularly in the fields of geometry, engineering, and physics. They relate the angles of a triangle to the lengths of its sides and have various applications across different scientific domains.



```python
x_coords = [0, 1, 0, 0]
y_coords = [0, 0, 1, 0]
plt.plot(x_coords, y_coords, 'k') 
plt.text(0.5, 0.5, "Hypotenuse",fontsize=15)
plt.text(0.01, 0.4, "opposite", fontsize=12)
plt.text(0.3, 0.03, "adjacent",fontsize=12 )
```

### Key Trigonometric Functions

### 1. Sine (sin)

The sine of an angle in a right triangle is defined as the ratio of the length of the side opposite the angle to the hypotenuse.

#### Formula:
$ \sin(\theta) = \frac{\text{opposite side}}{\text{hypotenuse}} $

### 2. Cosine (cos)

Cosine is the ratio of the length of the adjacent side to the hypotenuse in a right triangle.

#### Formula:
$ \cos(\theta) = \frac{\text{adjacent side}}{\text{hypotenuse}} $

### 3. Tangent (tan)

Tangent is the ratio of the sine of an angle to the cosine of that angle, equivalent to the ratio of the opposite side to the adjacent side.

#### Formula:
$ \tan(\theta) = \frac{\sin(\theta)}{\cos(\theta)} = \frac{\text{opposite side}}{\text{adjacent side}} $


```python
X=np.linspace(-5 * np.pi, 5 * np.pi, 200)
y = np.sin(X)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(X, y, "#AF8610", label= "sin(x)")
plt.title("Graph of Sin(X) function")
plt.legend(loc=4)
```

```python
X=np.linspace(-5 * np.pi, 5 * np.pi, 200)
y = np.cos(X)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(X, y, "#0A9EA4", label= "cos(x)")
plt.title("Graph of Cos(X) function")
plt.legend(loc=4)
```

```python
X=np.linspace(-5 * np.pi, 5 * np.pi, 200)
y = np.tan(X)
plt.figure(figsize=(10,8))
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(X, y, "r.", markersize=4 ,label= "Tan(x)")
plt.title("Graph of Tan(x) function")
plt.axis([-15,15,-20,20])
plt.legend(loc=4)
```

## 7. Hyperbolic Functions

Hyperbolic functions are analogs of trigonometric functions but are based on hyperbolas instead of circles.

## Key Hyperbolic Functions

### 1. Hyperbolic Sin $ ( \sinh ) $
The hyperbolic sine function is defined as:
$
\sinh(x) = \frac{e^x - e^{-x}}{2}
$

#### Properties:
- **Odd function**: $ \sinh(-x) = -\sinh(x) $
- **Range**:  $(-\infty, \infty)$
- **Derivative**: $ \frac{d}{dx} \sinh(x) = \cosh(x) $

---

### 2. Hyperbolic Cos $( \cosh )$
The hyperbolic cosine function is defined as:
$
\cosh(x) = \frac{e^x + e^{-x}}{2}
$

#### Properties:
- **Even function**: $ \cosh(-x) = \cosh(x) $
- **Range**: $ [1, \infty] $ 
- **Derivative**: $ \frac{d}{dx} \cosh(x) = \sinh(x) $

---

### 3. Hyperbolic Tan $( \tanh )$
The hyperbolic tangent function is defined as:
$
\tanh(x) = \frac{\sinh(x)}{\cosh(x)} = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$

#### Properties:
- **Odd function**: $ \tanh(-x) = -\tanh(x) $
- **Range**: $ (-1, 1) $
- **Derivative**: $ \frac{d}{dx} \tanh(x) = 1 - \tanh^2(x) $



```python
x = np.linspace(-5, 5, 100)
y = np.sinh(x)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, "b", label="$\\sinh(x)$")
plt.legend(loc=4)
plt.title("Graph of $\\sinh(x)$")
```

```python
x = np.linspace(-5, 5, 100)
y = np.cosh(x)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, "g", label="$\\cosh(x)$")
plt.legend(loc=4)
plt.title("Graph of $\\cosh(x)$")
```

```python
x = np.linspace(-5, 5, 100)
y = np.tanh(x)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, "r", label="$\\tanh(x)$")
plt.legend(loc=4)
plt.title("Graph of $ \\tanh(x) $")
```
